```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
credits = pd.read_csv('credits.csv')
```

```python
display(credits)
display(movies)
```

```python
movies = movies.merge(credits, on='title')
movies.shape
```

```python
movies.info()
```

```python
"""Select the columns that we need from the movies dataframe and read first 5 rows of the dataframe."""
movies = movies[['movie_id', 'title', 'genres', 'keywords', 'overview', 'cast', 'crew']]
movies.head()
```

```python
movies.info()
```

```python
"""Verify if there are any missing values in the dataset."""
movies.isnull().sum()
```

```python
"""Drop missing values from the dataset."""
movies.dropna(inplace= True)
```

```python
"""Verify if there are any duplicates in the data."""
movies.duplicated().sum()
```

```python
"""Read the first row of the movies DataFrame and extract the genres of the movie."""
movies.iloc[0].genres
```

```python
"""Import the ast - Abstract Syntax Tree - module to convert the stringified features to the list format."""
import ast
```

```python
"""Function that converts the stringified features to the list format."""
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
"""Converting the stringified features to the list format."""
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert_3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_3)
movies.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()  
```

```python
"""Separate the words in the overview column."""
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies.head()
```

```python
"""Removing the spaces between the words in the features."""
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])

movies.head()
```

```python
"""Puting all the features in the new tags column."""
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
"""Importing the CountVectorizer from the sklearn library
and creating an object of the CountVectorizer class 
that will convert the text data into token counts matrix."""
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
vector
```

```python
len(cv.get_feature_names_out())
```

```python
"""Importing the PorterStemmer from the nltk library to perform stemming on the text data."""
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
"""Function that performs stemming on the text data."""
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
"""cosine_similarity mesure the similarity between two non-zero vectors.
Mesure by the cosine of the angle between the two vectors.
That determines whether two vectors are pointing in roughly the same direction."""
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)
similarity
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
similarity[0].shape
```

```python
"""Create a sorted list. The enumerate() function converts a data collection object
in an enumarate object and returns an object that contains a counter as a key
for each value with an object. So the item within a collection is easier to access"""
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1])[1:6]
```

```python
def recommend_byMovie(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend_byMovie('The Matrix')
```
